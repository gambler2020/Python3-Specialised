```python
S="Hey this is regular Expression."
```

```python
x=re.split("\s", S) #this will split every single word by space
```

```python
#The split() function returns a list where the string has been split at each match.
```

```python
print(x)
```

```python
y=re.split("\s", S,1) #split occurred only on first word.
```

```python
print(y)
```

```python
z=re.sub("\s", "_", S) #replace white space character with _ or any other charecter you want.
```

```python
#The sub() function replaces the matches with the text of your choice
```

```python
print(z)
```
